```python
#importing necessary modules
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter, AutoMinorLocator)
```

```python
#defining variables
#R0 is the basic reproduction number, or how many people each infected person will infect
R0 = 1
population = 100
zStartingNum = 1
time = np.arange(0, 1000, 0.05)
if population <= 100:
    mult = 0.1
else:
    mult = 1

#the equation being used here is infected = startingnum * (R0 + 1)^time
infected = [zStartingNum * ((R0 + 1) ** t) for t in time]
humans = [population - i for i in infected]

ymin = 0
ymax = population
xmin = 0
xmax = np.interp(ymax + (mult * population), infected, time)
```

```python
fig, axes = plt.subplots(figsize = (7, 7))
axes.plot(time, infected, c="r", lw=3, ls="dashed")
axes.plot(time, humans, lw=2)

#setting major and minor ticks of graph 1
axes.xaxis.set_major_locator(MultipleLocator(round(xmax / 10)))
axes.xaxis.set_major_formatter(FormatStrFormatter('%d'))
if xmax > 20 and xmax <= 100:
    axes.xaxis.set_minor_locator(MultipleLocator(5))
axes.yaxis.set_major_locator(MultipleLocator(round(ymax / 5)))
axes.yaxis.set_major_formatter(FormatStrFormatter('%d'))
if ymax > 20 and ymax <= 100:
    axes.yaxis.set_minor_locator(MultipleLocator(5))

axes.set(ylim=(ymin, ymax), xlim=(xmin, xmax))
axes.set_xlabel("Time (Days)", fontsize=13)
axes.set_ylabel("Population", fontsize=13)
axes.set_title("Human and Zombie Poulations Over Time")

axes.grid(True)
```
